# 전처리


### lyricist & composer: drop

```python
DROP_COLS = ["lyricist", "composer"]
train.drop(columns=DROP_COLS, inplace=True)
test.drop(columns=DROP_COLS, inplace=True)
```

### 범주형 결측치 -> unknown으로 대체

총 8개

- gender

- source_screen_name

- source_system_tab

- source_type

- genre_ids

- artist_name

- name

- isrc

```python
cat_fill_cols = [
    "gender",
    "source_screen_name",
    "source_system_tab",
    "source_type",
    "genre_ids",
    "artist_name",
    "name",
    "isrc"
]

for c in cat_fill_cols:
    train[c] = train[c].fillna("unknown")
    test[c] = test[c].fillna("unknown")
```

### 수치형 결측치는 "값+결측여부"로


```python
# age

import numpy as np

train = train.rename(columns={"bd": "age"})
train["age"] = pd.to_numeric(train["age"], errors="coerce")
train.loc[(train["age"] < 0) | (train["age"] > 100), "age"] = np.nan
train["age_missing"] = train["age"].isna().astype("int8")

test = test.rename(columns={"bd": "age"})
test["age"] = pd.to_numeric(test["age"], errors="coerce")
test.loc[(test["age"] < 0) | (test["age"] > 100), "age"] = np.nan
test["age_missing"] = test["age"].isna().astype("int8")
```

```python
# song_length

train["song_length_missing"] = train["song_length"].isna().astype("int8")
train["song_length"] = train["song_length"].fillna(train["song_length"].median())

test["song_length_missing"] = test["song_length"].isna().astype("int8")
test["song_length"] = test["song_length"].fillna(test["song_length"].median())
```

### 날짜 컬럼 전처리 (문자 → 시간 정보)

- 20170115는 모델 입장에서 의미 없는 큰 숫자

- 우리는 “언제 가입했고, 얼마나 활동했는지”가 필요

```python
import pandas as pd

reg = pd.to_datetime(
    train["registration_init_time"].astype("Int64").astype(str),
    format="%Y%m%d",
    errors="coerce"
)

exp = pd.to_datetime(
    train["expiration_date"].astype("Int64").astype(str),
    format="%Y%m%d",
    errors="coerce"
)

train["reg_year"] = reg.dt.year.astype("Int16")
train["reg_month"] = reg.dt.month.astype("Int8")

train["exp_year"] = exp.dt.year.astype("Int16")
train["exp_month"] = exp.dt.month.astype("Int8")

train["days_active"] = (exp - reg).dt.days
train["days_active_missing"] = train["days_active"].isna().astype("int8")
```

```python
import pandas as pd

reg = pd.to_datetime(
    test["registration_init_time"].astype("Int64").astype(str),
    format="%Y%m%d",
    errors="coerce"
)

exp = pd.to_datetime(
    test["expiration_date"].astype("Int64").astype(str),
    format="%Y%m%d",
    errors="coerce"
)

test["reg_year"] = reg.dt.year.astype("Int16")
test["reg_month"] = reg.dt.month.astype("Int8")

test["exp_year"] = exp.dt.year.astype("Int16")
test["exp_month"] = exp.dt.month.astype("Int8")

test["days_active"] = (exp - reg).dt.days
test["days_active_missing"] = test["days_active"].isna().astype("int8")
```

### Count Encoding 피처 3개 만들기

(1) msno_cnt: 유저가 train에 얼마나 등장했나

(2) song_id_cnt: 노래가 얼마나 등장했나(인기)

(3) artist_cnt: 아티스트가 얼마나 등장했나

```python
import numpy as np
import pandas as pd

def add_count_features(train, test, cols):
    # train+test를 합쳐서 count를 구하면 "전체 데이터에서의 빈도"가 됨
    all_df = pd.concat([train[cols], test[cols]], axis=0, ignore_index=True)

    for c in cols:
        vc = all_df[c].value_counts(dropna=False)
        train[f"{c}_cnt"] = train[c].map(vc).astype("int32")
        test[f"{c}_cnt"]  = test[c].map(vc).astype("int32")

    return train, test

count_cols = ["msno", "song_id", "artist_name"]
train, test = add_count_features(train, test, count_cols)
```

추가로 “조금 더 좋은” count 피처 2개 (선택, 추천)

source 조합 빈도(어떤 경로로 많이 듣는지)

source_system_tab, source_screen_name, source_type를 합쳐서 “경로 ID” 만들고 count

```python
for df in [train, test]:
    df["source_combo"] = (
        df["source_system_tab"].astype(str) + "|" +
        df["source_screen_name"].astype(str) + "|" +
        df["source_type"].astype(str)
    ).astype("category")

train, test = add_count_features(train, test, ["source_combo"])
```

### count encoding 결과 해석

1. **msno_cnt (유저 활동량)**: median(50%) ≈ 78/ max ≈ 666

-> 유저마다 활동량 차이 있음

-> “헤비 유저 vs 라이트 유저” 구분 가능

-> 좋은 피처

2. **song_id_cnt (곡 인기)**: median ≈ 67/max ≈ 2030

->  인기곡 vs 비인기곡 명확

-> 추천 문제에서 가장 강력한 피처 중 하나

3. artist_name_cnt (아티스트 인기): median ≈ 1,388/max ≈ 36,930

-> 인기 아티스트 효과 아주 큼

-> 단, 분포가 심하게 오른쪽 꼬리(long-tail)

4. source_combo_cnt (경로 빈도): median ≈ 126,133/max ≈ 251,222

-> 특정 추천 경로가 엄청 자주 쓰임

-> 이건 행동 맥락 정보라서 굉장히 좋음

-> 다만 스케일이 너무 큼

```python
# 로그변환

import numpy as np

log_cols = [
    "msno_cnt",
    "song_id_cnt",
    "artist_name_cnt",
    "source_combo_cnt"
]

for c in log_cols:
    train[c + "_log"] = np.log1p(train[c])
    test[c + "_log"]  = np.log1p(test[c])
```
