# 2.1 파이토치 개요

## 2.1 파이토치 특징 및 장점
- GPU에서 텐서 조작 및 동적 신경망 구축이 가능한 프레임워크

image.png

- 파이토치는 효율적인 계산， 낮은 CPU 활용 직관적인 인터페이스와 낮은 진입 장벽 등을 장점으로 가지고 있음

## 2.1.2 파이토치의 아키텍처
- 가장 상위 계층은 파이토치 API 가 위치해 있으며 그 아래에는 파이토치 엔진
- 파이토치 엔진에서는 다차원 텐서 및 자동 미분을 처리
- 마지막으로 가장 아래에는 텐서에 대한 연산을 처리

### 파이토치 API

- 사용자 인터페이스를 제공하지만 실제 계산은 수행하지 않음

1. torch: GPU를 지원하는 텐서 패키지
2. torch.autograd: 자동 미분 패키지
3. torch.nn: 신경망 구축 및 훈련 패키지
4. torch.multiprocessing: 파이씬 멀티프로세싱 패키지
5. torch.utils: DataLoader 및 기타 유틸리티를 제공하는 패키지

### 파이토치 엔진

- Autograd C++, Aten C++, JIT C+十， Python API

### 연산 처리
- CPU와 GPU(TH(토치)， THC(토치 CUDA))를 이용하여 효율적인 데이터 구조 다차원 텐서에 대한 연산을 처리

### 텐서 생성 및 변환

```python
import torch
print(torch.tensor([[1,2], [3,4]]))
print(torch.tensor([[1,2], [3,4]], device="cpu")) # GPU에 텐서 생성
print(torch.tensor([[1,2], [3,4]], dtype=torch.float64))
```

```python
temp = torch.tensor([[1,2], [3,4]])
print(temp.numpy()) # 텐서를 ndarray로 변환

temp = torch.tensor([[1,2], [3,4]], device="cpu") # CPU에 텐서 생성
print(temp.numpy()) # CPU상의 텐서를 ndarray로 변환
```

### 텐서의 인덱스 조작

```python
temp = torch.FloatTensor([1,2,3,4,5,6,7]) # 파이토치로 1차원 벡터 생성
print(temp[0], temp[1], temp[-1]) # 인덱스로 접근
print('----------------------')
print(temp[2:5], temp[4:-1]) # 슬라이스로 접근
```

### 텐서 연산 및 차원 조작

```python
v = torch.tensor([1,2,3]) # 길이가 3인 벡터 생성
w = torch.tensor([3,4,6])
print(w - v) # 길이가 같은 벡터 간 뺄셈 연산
```

```python
#텐서의 차원 변경
temp = torch.tensor([
    [1,2], [3,4]]) # 2x2 행렬 생성

print(temp.shape)
print('-----------------')
print(temp.view(4,1)) # 4x1 행렬로 변형
print('-----------------')
print(temp.view(-1)) # 1차원 벡터로 변형
print('-----------------')
print(temp.view(1,-1)) # (1,4)
print('-----------------')
print(temp.view(-1,1)) # (4,1)
```
