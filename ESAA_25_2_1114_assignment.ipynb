## 5.1 합성곱 신경망

- 합성곱 신경망은 이미지 전체를 한 번에 계산하는 것이 아닌 이미지의 국소적 부분을 계산함으로써 시간과 자원을 절약하여 이미지의 세밀한 부분까지 분석할 수 있는 신경망

### 5.1.1 합성곱층의 필요성

- 이미지를 펼쳐서 분석하면 데이터의 공간적 구조를 무시하게 되는데， 이것을 방지하려고 도입된 것이 합성곱층임

1. 입력층
2. 합성곱층
3. 풀링층
4. 완전연결층
5. 출력층

#### 입력층

- 입력 이미지 데이터가 최초로 거치게 되는 계층

#### 합성곱층

- 입력 데이터에서 특성을 추출하는 역할을 수행

1단계. 입력 이미지에 3x3 필터 적용
2단계. 필터가 만큼 이동
3단계. 필터가 만큼 두 번째 이동
4단계. 필터가 만큼 세 번째 이동
5단계. 필터가 만큼 네 번째 이동
6단계. 필터가 만큼 마지막으로 이동

#### 풀링층

- 합성곱층과 유사하게 특성 뱀의 차원을 다운 샘플링하여 연산량을 감소시키고, 주요한 특성 벡터를 추출하고 학습을 효과적으로 할 수 있게 함

1. 최대 풀링: 대상 영역에서 최댓값 추출
2. 평균 풀링: 대상 영역에서 평균을 반환
 -> 대부분의 합성곱 신경망에서는 최대 풀링이 사용되는데 평균 풀링은 각 커널 값을 평균화시켜 중요한 가중치를 갖는 값의 특성이 희미해질 수 있기 때문


#### 완전 연결층

- 합성곱층과 풀링층을 거치면서 차원이 축소된 특성 뱀은 최종적으로 완전연결층으로 전달

#### 출력층

- 소프트맥스 활성화 함수가 사용되는데， 입력받은 값을 0~1 사이의 값으로 출력
